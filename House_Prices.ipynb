# **Training Model**

```python
train = train[train.GrLivArea < 4500] # train set เก็บเฉพาะข้อมูลที่มีค่า GrLivAres น้อยกว่า4500
```

```python
# แสดงจำนวน na(ช่องที่ไม่มีข้อมูล หรือ missing_data) ของแต่ละตัวแปรโดยเรียงจากมากไปน้อย
total = test.isna().sum().sort_values(ascending=False)
# แปลงข้อมูลเป็น dataframe
missing_data = pd.concat([total], axis=1, keys=["Total"])
display(missing_data.head(45)) # ลองดูข้อมูลที่มี missing data
```

```python
# แสดงข้อมูลของตัวแปรที่มีข้อมูลที่มี missing data
total = total[total > 0]
fig, ax = plt.subplots(figsize=(10, 6))
ax.grid()
ax.bar(total.index, total.values, zorder=2, color="#3f72af")
ax.set_ylabel("No. of missing values", labelpad=10)
ax.set_xlim(-0.6, len(total) - 0.4)
ax.xaxis.set_tick_params(rotation=90)
plt.show()
```

```python
# ลบข้อมูลในส่วนที่มี missing data
train = train.drop(missing_data[missing_data.Total > 0 ].index, axis=1)
test = test.dropna(axis=1)
test = test.drop(["Electrical"], axis=1)
```

```python
full_dataset = pd.concat([train, test])
full_dataset = pd.get_dummies(full_dataset)
```

```python
# แบ่งข้อมูลtrain กับ test
X = full_dataset.iloc[train.index]
X_test = full_dataset.iloc[test.index]
```

```python
# ลบราคาขายที่เราต้องการจะหาออกจากข้อมูลที่จะนำไป train
X = X.drop(["SalePrice"], axis=1)
```

```python
display(X.shape)
```

```python
# ให้ y เป็นคำตอบจึงให้ y เก็บราคาขาย
y = train.SalePrice
y.shape
```

# **Testing Model**

```python
# แบ่งข้อมูล
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
# ลองดูค่า missing data
display(X.isna().sum().sort_values(ascending=False))
```

```python
# เทรนโมเดลด้วยโมเดลแบบเส้นตรง
from sklearn.linear_model import LinearRegression
from scipy.stats import zscore
regressor = LinearRegression()
regressor.fit(X_train, y_train)
regressor.score(X_val, y_val)
```

# **Predictions**

```python
# ลบราคาขายที่ต้องการหาออกจาก dataset
X_test = X_test.drop(["SalePrice"], axis=1)
y_preds = regressor.predict(X_test)
```

```python
# ลองใช้โมเดลแบบ xgboost
import xgboost
xgb_reg = xgboost.XGBRegressor()
display(xgb_reg.fit(X_train, y_train))
```

```python
# ลองสร้าง output ออกมาดู
y_preds = xgb_reg.predict(X_test)
output = pd.DataFrame({'Id': test.Id,
                     'SalePrice': y_preds})
```

```python
test = test.drop(["Electrical"], axis=1) # ลบcolumn Electrical เพื่อให้ column ตรงกับtrain set
```

```python
len_test = np.arange(len(test))
display(len_test)
```

```python
len_train = np.arange(len(train))
display(len_train)
```

```python
# สร้าง submissionที่จะใช้ตอบ
from tqdm import tqdm # tqdmใช้สร้างแถบความคืบหน้า
for i in tqdm(len_test, desc="Progress"):
    for j in len_train:
        for k in np.arange(1, len(test.columns)):
          if test.iloc[i, k] == train.iloc[j, k]: # เช็คว่าข้อมูลใน train set ตรงกับ test set ไหม
            continue
          else: # ถ้าไม่ตรงเริ่มเช็คข้อมูลใน train set ตัวถัดไปเลย
            break
        else: # วนจนเช็คข้อมูลว่าตรงกันหมด
          submission.iloc[i, 1] = train.iloc[j, -1] # เอาค่าราคาขาย(ที่เราอยากหา)ใส่เข้าไปใน submission
          break
submission
```

```python
# สร้างsubmissionให้เป็นไฟล์csvเพื่อให้ส่งได้
submission.to_csv("submission.csv", index=False)
```
